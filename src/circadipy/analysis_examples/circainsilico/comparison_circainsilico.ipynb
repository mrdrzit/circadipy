```python
### IMPORT LIBRARIES ----------------------------------------------------------------------------------------------------

%matplotlib qt

%reload_ext autoreload
%autoreload 3

import sys                                                                                                              # Import sys to add paths to libraries
import matplotlib.pyplot as plt                                                                                         # Import matplotlib.pyplot to plot figures
import tkinter as tk                                                                                                    # Import TK to open folder dialogs to select folders
from tkinter import filedialog                                                                                          # Import filedialog to open folder dialogs to select folders
import numpy                                                                                                            # Import numpy to work with arrays and make calculations
from shutil import rmtree
from datetime import datetime, timedelta                                                                                # Import time to measure time 
import time                                                                                                             # Import time to measure time
import os                                                                                                               # Import path to work with paths
import pandas                                                                                                           # Import pandas to work with dataframes
import warnings                                                                                                         # Import warnings to ignore warnings
warnings.filterwarnings('ignore')                                                                                       # Ignore warnings

## IMPORT CIRCADIPY ----------------------------------------------------------------------------------------------------

sys.path.append('E:\\github\\nnc-ufmg\\circadipy\\src')
import chrono_reader as chr                                                                                             # Import chrono_reader to read data
import chrono_plotter as chp                                                                                            # Import chrono_plotter to plot data
import chrono_rythm as chrt                                                                                             # Import chrono_rithm to make calculations
import chrono_simulation as chs                                                                                         # Import chrono_simulation to simulate data
```

```python
## SET THE ENVIRONMENT -------------------------------------------------------------------------------------------------

root = tk.Tk()                                                                                                          # Create the root window
root.attributes('-topmost',True)                                                                                        # Keep the window on top of others
root.iconify()                                                                                                          # Hide the root window

database_file = filedialog.askopenfilename(initialdir = os.getcwd(), title = "Select database file")                    # Ask the user to select the database file

main_save_folder = os.path.dirname(database_file)                                                                       # Get the folder where the database file is located
root.destroy()                                                                                                          # Destroy the root window

print('The database file is located in: ' + main_save_folder)                                                           # Print the folder where the database file is located
```

```python
def reset_folder(folder):                                                                                               # Function to reset the folder
    for root, dirs, files in os.walk(folder):                                                                           # For each file in the folder
        for f in files:                                                                                                 # For each file
            os.unlink(os.path.join(root, f))                                                                            # Remove the file
        for d in dirs:                                                                                                  # For each folder
            rmtree(os.path.join(root, d))                                                                               # Remove the folder
```

```python
## READ THE DATABASE ---------------------------------------------------------------------------------------------------

remove_file = False                                                                                                     # If the file should be removed

database = {}                                                                                                           # Create a dictionary with the database
save_folder = main_save_folder + '/data_circainsilico/'                                                                 # Define the folder to save the data
result_folder = main_save_folder + '/results_circainsilico/'                                                            # Define the folder to save the results

## CREATE THE FOLDERS --------------------------------------------------------------------------------------------------

if not os.path.exists(save_folder):
    os.makedirs(save_folder)
else:
    if remove_file:                                                                                                     # If the file should be removed
        reset_folder(save_folder)                                                                                       # Reset the folder
        print("Data folder reseted")                                                                                    # Print that the folder was reseted
    else:                                                                                                               # If the file shouldn't be removed
        print("Data folder already exists")                                                                             # Print that the folder already exists

if not os.path.exists(result_folder):
    os.makedirs(result_folder)
else:
    if remove_file:                                                                                                     # If the file should be removed
        reset_folder(result_folder)                                                                                     # Reset the folder
        print("Result folder reseted")                                                                                  # Print that the folder was reseted
    else:                                                                                                               # If the file shouldn't be removed
        print("Result folder already exists")                                                                           # Print that the folder already exists

## READ THE DATABASE ---------------------------------------------------------------------------------------------------

database_df = pandas.read_csv(database_file)                                                                            # Read the database file
display(database_df.head())                                                                                             # Display the database

for row in database_df.iterrows():
    database[row[0]] = {}
    database[row[0]]['Time'] = range(1, 49)
    database[row[0]]['Values'] = row[1][1:-3].to_list()
    database[row[0]]['Amplitude'] = row[1]['Cosine Amplitude']
    database[row[0]]['Period'] = row[1]['Cosine Periods']
    database[row[0]]['Phase'] = row[1]['Cosine Lag Factors']

print(database)
```

```python
## CREATE A GENERIC ASC TO CIRCADIPY -----------------------------------------------------------------------------------

# Set a generic date with the format MM/DD/YY,HH:MM:SS
generic_date = '01/01/20,00:00:00'
generic_date = datetime.strptime(generic_date, '%m/%d/%y,%H:%M:%S')

# Set the generic time with the format HH:MM:SS with the times (in hours) in the database
for id in database:     
    database[id]['circadipy'] = [generic_date + timedelta(hours=t) for t in database[id]['Time']]
    database[id]['circadipy'] = [datetime.strftime(t, '%m/%d/%y,%H:%M:%S') for t in database[id]['circadipy']]
    database[id]['circadipy'] = [t + ',' + str(v) for t, v in zip(database[id]['circadipy'], database[id]['Values'])]
    file_name = 'simulation_' + str(id) + '.asc'
    file_to_save = os.path.join(save_folder, file_name)
    database[id]['file_name'] = file_to_save
    with open(file_to_save, 'w') as f:
        f.write('\n'.join(database[id]['circadipy']))
    f.close()
```

```python
## CONFIGURE THE EXPERIMENT --------------------------------------------------------------------------------------------

zt_0_time = 0                                                                                                           # Set the time for the first ZT (0)
type = 'generic'                                                                                                        # Set the type of the data decoding (endocrino or intellicage)
labels_dict = {'cycle_types': ['LD'], 'test_labels': ['Gene'], 'cycle_days': []}                                        # Create a dictionary to store to pass as argument to the read_protocol function

## READ THE PROTOCOL ---------------------------------------------------------------------------------------------------

for id in database:     
    file = database[id]['file_name']
    name = 'simulation_' + str(id)
    database[id]['protocol'] = chr.read_protocol(name, file, zt_0_time = zt_0_time, labels_dict = labels_dict, 
                                                 type = type, consider_first_day = True)                                # Read the protocol (chrono_reader.py)
    database[id]['protocol'].apply_filter(window = 3, type = 'moving_average', order = 2, reverse = False)              # Apply a filter to the data (chrono_reader.py)
```

```python
## SET THE PARAMETERS FOR THE COSINOR FITTING --------------------------------------------------------------------------

dict = {'time_shape': 'continuous',                     
        'step': 0.01, 
        'start_time': 20, 
        'end_time': 28, 
        'n_components': [1]}                                                                                            # Create a dictionary to pass as argument to the fit_cosinor function

## FIT THE COSINOR TO THE DATA -----------------------------------------------------------------------------------------

init = time.time()                                                                                                      # Get the initial time

for id in database:     
    protocol = database[id]['protocol']
    best_models, _ = chrt.fit_cosinor(protocol, dict = dict, save_folder = result_folder)                               # Fit the cosinor to the data (chrono_rithm.py)
    database[id]['best_models'] = best_models
    chp.model_over_signal(protocol, best_models, position = 'head', mv_avg_window = 1, save_folder = result_folder,
                          format = 'svg')                                                                               # Plot representative figure with the data and the respective model (using chrono_plotter.py)

end = time.time() - init                                                                                                # Get the time elapsed
```

```python
## COMPARE THE COSINOR FITTING WITH THE GORUND-TRUTH PERIODS -----------------------------------------------------------

period_error = []                                                                                                       # Create an empty list to store the model error
phase_error = []                                                                                                        # Create an empty list to store the model error
amplitude_error = []                                                                                                    # Create an empty list to store the model error
comparison_file = result_folder + 'comparison_ciicainsilico.txt'                                                        # Create the path to the file with the comparison results

for id in database:         
    name = 'simulation_' + str(id)                                                                                      # Get the gene name

    period_circadipy = database[id]['best_models']['period'][0]                                                         # Get the activity period calculated by the model
    period_other = database[id]['Period']                                                                               # Get the activity period used to generate the data
    
    phase_circadipy = database[id]['best_models']['acrophase_zt'][0]                                                    # Get the activity phase calculated by the model
    phase_other = database[id]['Phase']*period_other                                                                    # Get the activity phase used to generate the data

    amplitude_circadipy = database[id]['best_models']['amplitude'][0]                                                   # Get the p-value calculated by the model
    amplitude_other = database[id]['Amplitude']                                                                         # Get the p-value used to generate the data

    period_error.append(period_circadipy - period_other)                                                                # Calculate the model error for the period    
    phase_error.append(phase_circadipy - phase_other)                                                                   # Calculate the model error for the phase
    amplitude_error.append(amplitude_circadipy - amplitude_other)                                                       # Calculate the model error for the p-value

    if phase_error[-1] > 12:
        phase_circadipy -= period_circadipy
        phase_error[-1] = phase_circadipy - phase_other
    elif phase_error[-1] < -12:
        phase_other -= period_other
        phase_error[-1] = phase_circadipy - phase_other

    with open(comparison_file, 'a') as f:                                                                               # Open the file to store the comparison results
        f.write('\n------------------------------------------------------------------')
        f.write('\nGene: ' + str(name))
        f.write('\nCircadiPy: ' + str(period_circadipy) + ' Other: ' + str(period_other))                                 # Print the period calculated by the model and the ground-truth period
        f.write('\nThe period error for ' + str(id) + ' is: ' + str(period_error[-1]))                                    # Print the model error for the period
        f.write('\nCircadiPy: ' + str(phase_circadipy) + ' Other: ' + str(phase_other))                                   # Print the phase calculated by the model and the ground-truth phase
        f.write('\nThe phase error for ' + str(id) + ' is: ' + str(phase_error[-1]))                                      # Print the model error for the phase
        f.write('\nCircadiPy: ' + str(amplitude_circadipy) + ' Other: ' + str(amplitude_other))                           # Print the p-value calculated by the model and the ground-truth p-value
        f.write('\nThe p-value error for ' + str(id) + ' is: ' + str(amplitude_error[-1]))                                # Print the model error for the p-value
    f.close()                                                                                                           # Close the file

period_error = numpy.array(period_error)                                                                                # Convert the list to a numpy array
phase_error = numpy.array(phase_error)                                                                                  # Convert the list to a numpy array
amplitude_error = numpy.array(amplitude_error)                                                                          # Convert the list to a numpy array

mean_period_error = numpy.mean(period_error)                                                                            # Calculate the mean error for the period
mean_phase_error = numpy.mean(phase_error)                                                                              # Calculate the mean error for the phase
mean_amplitude_error = numpy.mean(amplitude_error)                                                                      # Calculate the mean error for the p-value

std_period_error = numpy.std(period_error)                                                                              # Calculate the standard deviation for the period
std_phase_error = numpy.std(phase_error)                                                                                # Calculate the standard deviation for the phase
std_amplitude_error = numpy.std(amplitude_error)                                                                        # Calculate the standard deviation for the p-value

with open(comparison_file, 'a') as f:                                                                                   # Open the file to store the comparison results
    f.write('\n==================================================================')
    f.write('\nThe mean error for the period parametrization is: ' + str(round(mean_period_error, 4)) + ' +/- ' + \
        str(round(std_period_error, 4)))                                                                                # Print the mean error for the period
    f.write('\nThe mean error for the phase parametrization is: ' + str(round(mean_phase_error, 4)) + ' +/- ' + \
        str(round(std_phase_error, 4)))                                                                                 # Print the mean error for the phase
    f.write('\nThe mean error for the p-value parametrization is: ' + str(round(mean_amplitude_error, 4)) + ' +/- ' + \
        str(round(std_amplitude_error, 4)))                                                                             # Print the mean error for the p-value
f.close()
```

```python
## PLOT HISTOGRAM TO PERIOD, PHASE AND AMPLITUDE -----------------------------------------------------------------------

fig = plt.figure(figsize=(9, 3))
fig.suptitle('Parametrization Error (CircadiPy - CircaInSilica Ground Truth)', fontsize=16)
ax1 = fig.add_subplot(1, 3, 1)
ax1.hist(period_error, bins=32, color='midnightblue', edgecolor='midnightblue')
ax1.set_title('PERIOD\nMEAN: ' + str(round(mean_period_error,2)) + ' +/- ' + str(round(std_period_error,2)), fontsize=16)
ax1.set_xlabel('PERIOD (HOUR)', fontsize=16)
ax1.set_ylabel('FREQUENCY', fontsize=16)
ax1.set_xlim(-3.5, 3.5)
ax2 = fig.add_subplot(1, 3, 2)
ax2.hist(phase_error, bins=32, color='midnightblue', edgecolor='midnightblue')
ax2.set_title('PHASE\nMEAN: ' + str(round(mean_phase_error,2)) + ' +/- ' + str(round(std_phase_error,2)), fontsize=16)
ax2.set_xlabel('PHASE (HOUR)', fontsize=16)
ax2.set_xlim(-3.5, 3.5)
ax3 = fig.add_subplot(1, 3, 3)
ax3.hist(amplitude_error, bins=32, color='midnightblue', edgecolor='midnightblue')
ax3.set_title('AMPLITUDE\nMEAN: ' + str(round(mean_amplitude_error,2)) + ' +/- ' + str(round(std_amplitude_error,2)), fontsize=16)
ax3.set_xlabel('AMPLITUDE', fontsize=16)
ax3.set_xlim(-1,1)
plt.tight_layout()

plt.savefig(result_folder + '/parametrization_error_ground_truth.svg', dpi=300, format='svg')
plt.close()
```
